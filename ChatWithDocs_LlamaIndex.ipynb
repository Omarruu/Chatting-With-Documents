```python
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, ServiceContext
from llama_index.core import StorageContext, load_index_from_storage
from llama_index.core import Settings
from langchain.embeddings.huggingface import HuggingFaceBgeEmbeddings
from llama_index.llms.gemini import Gemini
import os
```

**Load Data**

```python
!mkdir data
```

```python
documents = SimpleDirectoryReader("data").load_data()
```

```python
documents
```

**Split text into small chunks**

```python
os.environ['GOOGLE_API_KEY'] = ''
```

```python
llm = Gemini(
    model="models/gemini-2.0-flash",        # or whichever gemini variant you prefer
    # api_key=os.environ["GOOGLE_API_KEY"]  # uses the env var by default
)

# 4) Embeddings (unchanged)
embed_model = HuggingFaceBgeEmbeddings(model_name="BAAI/bge-base-en")
```

```python
#service context is same as text splitter in langchain

Settings.llm = llm
Settings.embed_model = embed_model
Settings.chunk_size = 100
Settings.chunk_overlap = 20
```

```python
index = VectorStoreIndex.from_documents(documents)
```

**Storing and Loading the index**

```python
index.storage_context.persist()
```
